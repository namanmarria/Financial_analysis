```python
ms['MA10'] = ms['Close'].rolling(10).mean() # fast signal
ms['MA50'] = ms['Close'].rolling(50).mean() # slow signal
ms.head()
```

```python
plt.figure(figsize=(20,16))
ms['Close'].plot(label='Close')
ms['MA10'].plot(label='Fast Signal')
ms['MA50'].plot(label='Slow Signal')
# plotting

plt.legend()
plt.show()
```

```python
# create new variable Long or NOT ( Long- BUY AND HOLD THE STOCK)
ms['Shares'] = [1 if ms.loc[ei,'MA10'] > ms.loc[ei,'MA50'] else 0 for ei in ms.index]
ms
```

```python
# Daily Profit
ms['Close1'] = ms['Close'].shift(-1)
```

```python
ms.head()
```

```python
#Daily Profit
ms['Profit'] = [ms.loc[ei,'Close1']-ms.loc[ei,'Close'] if ms.loc[ei,'Shares'] == 1 else 0 for ei in ms.index]
```

```python
ms.tail()
```

```python
plt.figure(figsize=(20,16))
ms['Profit'].plot(legend=True)
# mark a border
plt.axhline(y=0,color="red")
```

```python
ms['wealth'] = ms['Profit'].cumsum()
ms.tail()
```

```python
# Cumulative wealth
print('Total money earned is ',ms.loc[ms.index[-2],'wealth'])
print('Total money you spent is ',ms.loc[ms.index[0],'Close'])
plt.figure(figsize=(20,16))
ms['wealth'].plot(legend=True)
```
